## 5. Load Model (Qwen3-1.7B, 4-bit)

Loads the base model with Unsloth optimizations and attaches a LoRA adapter (r=16).

```python
import sys, os, torch, gc, shutil

sys.path.insert(0, "scripts")

# Clear stale unsloth compiled cache
cache_dir = os.path.expanduser("~/astrof/unsloth_compiled_cache")
if os.path.exists(cache_dir):
    shutil.rmtree(cache_dir)
    print("Cleared unsloth compiled cache")

from unsloth import FastLanguageModel

MODEL_NAME = os.environ["MODEL_NAME"]

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=MODEL_NAME,
    max_seq_length=2048,
    dtype=None,
    load_in_4bit=True,
)

model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=2026,
)

print(f"Model loaded: {MODEL_NAME}")
print(
    f"Trainable params: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}"
)

# Report actual model dtype so we know which precision to use
_dtype = next(p for p in model.parameters() if not p.requires_grad).dtype
print(f"Base model dtype: {_dtype}")
```

## 6. SFT Warm-start

Fine-tunes one epoch on 5,780 greedy-policy demonstrations (`data/sft_warmstart.jsonl`).
This primes the model to output valid JSON for all 5 role-conditioned prompts before RL begins.
Without this, GRPO sees only zero-reward rollouts in early training.

**Expected time:** ~10 min on A100

```python
import torch
import train_grpo as tg

# Detect model dtype and patch SFTConfig + GRPOConfig to match — prevents
# Unsloth's fp16/bf16 mismatch error regardless of GPU type.
_model_dtype = next(p for p in model.parameters() if not p.requires_grad).dtype
_use_fp16 = _model_dtype == torch.float16
_use_bf16 = _model_dtype == torch.bfloat16
print(f"Model dtype: {_model_dtype} → fp16={_use_fp16}, bf16={_use_bf16}")

from trl import SFTConfig, GRPOConfig
import functools

# Unwrap any previous patches to avoid recursion
_orig_sft = getattr(SFTConfig.__init__, "__wrapped__", SFTConfig.__init__)
_orig_grpo = getattr(GRPOConfig.__init__, "__wrapped__", GRPOConfig.__init__)


@functools.wraps(_orig_sft)
def _sft_patched(self, *a, **kw):
    kw["bf16"] = _use_bf16
    kw["fp16"] = _use_fp16
    _orig_sft(self, *a, **kw)


@functools.wraps(_orig_grpo)
def _grpo_patched(self, *a, **kw):
    kw["bf16"] = _use_bf16
    kw["fp16"] = _use_fp16
    kw["max_completion_length"] = 1024
    _orig_grpo(self, *a, **kw)


SFTConfig.__init__ = _sft_patched
GRPOConfig.__init__ = _grpo_patched
print("Patched SFTConfig + GRPOConfig: dtype-matched, max_completion_length=1024")

tg.sft_warmstart(model, tokenizer)
print("SFT warm-start complete.")
```

| 3 | hard | Weather recovery, transit deadlines |

**Expected time:** ~20 min per stage on A100 (100 steps each)

```python
import os
import sys

sys.path.insert(0, os.getcwd())
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

import scripts.train_grpo as tg

# patch ENV BEFORE calling train_grpo
tg.ENV_BASE_URL = "https://Cosmosatom-astrof.hf.space"
tg.GRPO_STEPS_PER_TASK = 10  # quick test
print("ENV:", tg.ENV_BASE_URL)
print("Steps:", tg.GRPO_STEPS_PER_TASK)

for task_id in tg.CURRICULUM:
    print(f"\n{'=' * 50}\nTraining on task: {task_id}\n{'=' * 50}")
    tg.train_grpo(model, tokenizer, task_id)
```

## 12. Benchmark — Before vs After

Run the benchmark script with the trained model to get your AFTER numbers.
Compare against the zero-shot baseline already collected.

**Requires:** Ollama running with the trained model, OR set `API_BASE_URL` to HF Inference.

```python
# Option A: run inference directly in this notebook using the trained model
import sys, os, json, re

sys.path.insert(0, ".")
sys.path.insert(0, "scripts")

from train_grpo import _make_direct_model_fn, _run_episode, MAX_STEPS_PER_TASK

model_fn = _make_direct_model_fn(model, tokenizer)

print("Running trained model evaluation...")
results = {}
for task in ["easy", "medium", "hard", "expert"]:
    steps = _run_episode(task, model_fn)
    rewards = [s["reward"] for s in steps]
    avg_reward = sum(rewards) / max(len(rewards), 1)
    results[task] = round(avg_reward, 4)
    print(f"  {task:8s}  avg_reward={avg_reward:.4f}  steps={len(rewards)}")

print("\nTrained model results:")
print(json.dumps(results, indent=2))
```
